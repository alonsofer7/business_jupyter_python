```python
import pandas as pd
import numpy as np

df= pd.read_csv('sales_null.csv', encoding ="latin1")
df.head()
```

```python
df.dtypes
```

```python
#columns that have null values
df.isnull().sum()
```

```python
#Use fillna to fill in missing data with '0'.
df1 = df.fillna(0)
df1.head(3)
```

```python
# to change ALL Floating point numbers to 2 decimal places.
pd.options.display.float_format = '{:.2f}'.format
df1
```

```python
#to change the type of number to Integer (so no decimal places)
df1[['Units Sold']] = df1[['Units Sold']].astype(int)
df1
```

```python
#create a column.
df1['Total Profit']= df['Units Sold'] * df['Unit Price']
df1.head()
```

```python
total_profit = df1['Total Profit'].sum()
print('Total Profit =', total_profit)
```

```python
profits = df1['Total Profit'].sort_values(ascending=False)
print ('Total profits =', 
       profits)
```

```python
#To use groupby (by lowest Total Revenue first)
df1.groupby(['Region'])
print (df1.groupby('Region').groups)
```

```python
#To use groupby (by lowest Total Revenue first)
df1.groupby(['Region']).sum().sort_values('Total Revenue', ascending=True)
```

```python
#To create a Pivot Table
table = pd.pivot_table(df1, index=['Country','Sales Channel'], values=['Total Profit','Units Sold'],aggfunc=[np.sum], margins = True).round(2)
table
```

```python
#To create a Pivot Table. 'Index' has to be 'String'. 'Values' need to be 'int' or 'floating'.
profit_table = pd.pivot_table(df1, index=['Country'], values=['Total Profit','Units Sold'],aggfunc=[np.sum], margins = True).round(2)
profit_table
```

```python
#use query method
df1.query('Country ==["Albania"]')
```

```python
#Select data using .loc method. Alternative to .query method. 
df1.loc[(df1['Country'] == 'Australia')]
```

```python
#The .loc method allows for sorting of values.
df1.loc[(df1['Country'] == 'Australia'),['Country','Item Type','Total Profit','Units Sold']].sort_values("Total Profit", ascending=True).head(1)
```

```python
#Q6. Use .loc to find 'Australian' sales that is 'Offline' & Units sold > 5000. Then display only Country, Sales Channel & Units Sold
df1.loc[(df1['Country'] == 'Australia') & (df1['Sales Channel'] =='Offline') & (df1['Units Sold'] > 5000),['Country','Sales Channel','Units Sold']]  
```
