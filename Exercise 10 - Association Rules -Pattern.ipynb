```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv("groceries.csv")
data.head()
```

```python
data.shape
```

```python
#To get the Item names in the columns and each transaction (basket) to the rows.
basket = pd.get_dummies(data)
basket.head(1)
```

```python
#To create basket sets for all items
basket_sets = basket.drop("Transaction_number", axis = 1)
basket_sets.head(3)
```

```python
#Use Aprioir algorithm to build association rules. Go trough the basket set & extract the results
#to populate a Dataframe. Only number of items => 2% (support = 0.02) of total basket.
apriori(basket_sets, min_support=0.02).head()
```

```python
#To show item names in the basket sets that are popular the store (=>5%)
#Tail shows 5 itemsets with a total => 5% frequency
apriori(basket_sets, min_support=0.05, use_colnames= True).tail(2)
```

```python
#to show more items (reduce the min support level to 0.2) as well as the lenght of the itemset display.
df = basket_sets
frequent_itemsets = apriori(basket_sets, min_support=0.02, use_colnames=True)
#add a Lenght column to include combinations of all the item sets.
#so use code 'Lambda x:len' go to end fo the datase.
frequent_itemsets['lenght'] = frequent_itemsets['itemsets'].apply(lambda x:len(x))
frequent_itemsets.head()
```

```python
#to identify the item sets with a min basket size of 3 items.
frequent_itemsets[frequent_itemsets['lenght'] >=3] 
```
