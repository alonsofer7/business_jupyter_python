```python
import pandas as pd
import numpy as np
```

```python
#reading in a csv as a DataFrame (df).  .Head() method is to display only top 5 rows.

df = pd.read_csv('sales_null.csv', encoding='latin1')
df.head()
```

```python
#shows data types & names
df.dtypes
```

```python
#columns that have null values
df.isnull().sum()
```

```python
#Use fillna to fill in missing data with '0'.
df1 = df.fillna(0)
df1.head(2)
```

```python
# to change ALL Floating point numbers to 2 decimal places.
pd.options.display.float_format ='{:.2f}'.format
df1.head(2)
```

```python
#to change the type of number to Integer (so no decimal places)

df1['Units Sold'] = df1['Units Sold'].astype(int)
print ('Units Sold', df1['Units Sold'])
```

```python
df1.head(2)
```

```python
#Create a new calculated column called 'Total Profit'.
df1['Total Profit'] = df1['Units Sold'] * df1['Unit Price']
df1.head(1)
```

```python
#To create a sum of all the profits.

Total_Profit = df1['Total Profit'].sum()
print ('Total Profit =', Total_Profit)
```

```python
#To sort the values Highest to Lowest in Total Profit.
profits = df1['Total Profit'].sort_values(ascending=False)
print ('Total Profits =', profits) 
```

```python
#To use groupby
df1.groupby(['Region'])
print (df1.groupby('Region').groups)
```

```python
#To use groupby to sum all the Regions by Total Revenue (by Highest Total Revenue first)
df1.groupby(['Region']).sum().sort_values('Total Revenue',ascending=False).head(2)
```

```python
#To use groupby to sum all the Regions by Total Revenue (by Highest Total Revenue first)
Revenue_Table = df1.groupby(['Region']).sum().sort_values('Total Revenue',ascending=False)
Revenue_Table[['Total Revenue','Total Cost']]
```

```python
#To query method to except specific data.
Revenue_Table.query('Region == ["Asia"]') 
```
