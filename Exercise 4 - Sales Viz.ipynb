```python
#Seaborn library improves the look of graphs in matplotlib.
#Matplotlib creates inline plots.

import matplotlib.pyplot as plt
import seaborn
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
#use index to set the index to teh most appropriate column
df=pd.read_csv('sales_data.csv',encoding = "latin1")
df.set_index('COUNTRY').head(3)
```

```python
#to plot a range of x & y coordinates using matlib plot function
x = range(1,10)
y = [1,2,3,4,5,4,3,2,1]
plt.plot(x,y)
```

```python
#to create a line graph of 1 column of SALES
Sales = df['SALES']
Sales.plot()
```

```python
Q3. #Calculate new column 'Units Sold' and show only relevant columns for each order.
df['UNITSSOLD'] = df['SALES']/df['QUANTITYORDERED']
df[['ORDERNUMBER','UNITSSOLD','SALES']].head(3)
```

```python
#To create 2 columns of Units SOLD and SALES 
sold = df[['UNITSSOLD','SALES']]
sold.plot()
```

```python
#plot the difference in 'Products' according to 'SALES'
df_product = df[['PRODUCTLINE','SALES','UNITSSOLD']]
df_product.groupby(['PRODUCTLINE']).sum().plot(kind = 'bar')
```

```python
#plot the difference in 'COUNTRIES' according to 'SALES' - line chart.  2 columns in DF.
df_country = df[['COUNTRY','SALES']]
df_country.groupby(['COUNTRY']).sum().plot(kind = 'line')
```

```python
#plot a scatterplot for SAles & Quantity ORDERED.
df.plot(kind='scatter', x='SALES', y='QUANTITYORDERED', color='red')
plt.show()
```

```python
#Find sales >240
df_topsales = df.loc[(df['UNITSSOLD'] > 245),['CONTACTFIRSTNAME','UNITSSOLD','SALES']].head()
df_topsales
```

```python
#to locate the top sales people > 240 units sold
df_topsales.groupby('CONTACTFIRSTNAME').sum().sort_values('SALES', ascending=False).plot(kind='bar',color='blue')
```
